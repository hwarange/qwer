## 5.1 조건문

조건문은 특정 조건을 만족하는지 여부에 따라 실행하는 코드가 달라져야 할 때 사용.  
예를 들어 삼각형이 있을 시 삼각형 변의 길이에 따라  
정, 이등변, 직각, 일반 등의 삼각형으로 구분할 경우 조건이 필요.

### 5.1.1 if

파이썬은 이와 같은 상황에서 사용할 수 있는 if문 제공

사용법
* if 조건식:
<br>&nbsp;&nbsp;&nbsp;&nbsp;조건식의 결과가 True일 때 실행문

<br>
조건식의 결과가 True 이면 실행문이 동작하고,
조건식의 결과가 False면 실행문이 동작하지 않음.
<br>

**※주의**
파이썬은 타 언어와 다르게 코드 블록을 나타내는  
중괄호({})가 없는 대신 들여쓰기로 구분.  
따라서 다음과 같이 쓸 경우 오류 발생.

* if 조건식:
<br>조건식의 결과가 True일 때 실행문


들여쓰기 규칙은 다음과 같음  
1. 공백이나 탭을 이용하여 들여쓰기 수행  
2. 공백의 개수는 상관 없음  
그러나 파이썬 스타일 가이드(PEP 8)에 따르면 공백의 개수는 4개 사용.  
3. 탭은 1개만 사용  
4. 동일 구역에서 들여쓰기는 통일  
(탭과 공백을 혼용하여 사용할 수 없고, 들여쓰기 수준도 동일)

### 5.1.2 if-else

if-else문은 조건식을 만족하는 경우와 만족하지 않는 경우를 
구분할 필요가 있을 시 사용.

사용법
* if 조건식:
<br>&nbsp;&nbsp;&nbsp;&nbsp;조건식의 결과가 True일 때 실행문  
  else:
<br>&nbsp;&nbsp;&nbsp;&nbsp;조건식의 결과가 False일 때 실행문

조건식의 결과가 True인 경우와 False인 경우로 구분하여 코드 작성 가능.

### 5.1.3 if-elif

if-elif문은 특정 조건에 따라 3가지 이상으로 구분해야 할 때 사용.
여기서 elif는 else if

사용법
* if 조건식1:
<br>&nbsp;&nbsp;&nbsp;&nbsp;조건식1의 결과가 True일 때 실행문  
  elif 조건식2:
<br>&nbsp;&nbsp;&nbsp;&nbsp;조건식1의 결과가 False이고 조건식2의 결과가 True일 때 실행문.  
  elif 조건식3:
<br>&nbsp;&nbsp;&nbsp;&nbsp;조건식1, 2의 결과가 False이고 조건식 3의 결과가 True일 때 실행문

### 5.1.4 if-elif-else

if-elif문은 특정 조건에 따라 구분 후 위의 조건을 만족하는 않는  
모든 경우에 대해 실행을 할 때 사용.

사용법
* if 조건식1:
<br>&nbsp;&nbsp;&nbsp;&nbsp;조건식1의 결과가 True일 때 실행문  
  elif 조건식2:
<br>&nbsp;&nbsp;&nbsp;&nbsp;조건식1의 결과가 False이고 조건식2의 결과가 True일 때 실행문.  
  elif 조건식3:
<br>&nbsp;&nbsp;&nbsp;&nbsp;조건식1, 2의 결과가 False이고 조건식 3의 결과가 True일 때 실행문  
  else:
<br>&nbsp;&nbsp;&nbsp;&nbsp;조건식1, 2, 3이 모두 False일 때 실행될 문장.

```python
user_input = input('입력: ')
if (user_input == '안녕') or (user_input == '안녕하세요.') :
    print('안녕하세요.')
elif (user_input == '지금 몇 시야?') or (user_input == '지금 몇 시예요?'):
    print('지금은 15시 입니다.')
else:
    print(user_input)
```

```python
num = input('정수를 입력해주세요: ')

if num.isdecimal(): # 정수가 아닌 경우 아무것도 출력 안됨
    num = int(num)

    if num % 2 == 0 :
        print(f'{num}은 2로 나누어 떨어지는 숫자입니다.')
    else:
        print(f'{num}은 2로 나누어 떨어지는 숫자가 아닙니다.')

    if num % 3 == 0 :
        print(f'{num}은 3으로 나누어 떨어지는 숫자입니다.')
    else:
        print(f'{num}은 3으로 나누어 떨어지는 숫자가 아닙니다.')

    if num % 4 == 0 :
        print(f'{num}은 4로 나누어 떨어지는 숫자입니다.')
    else:
        print(f'{num}은 4로 나누어 떨어지는 숫자가 아닙니다.')

    if num % 5 == 0 :
        print(f'{num}은 5로 나누어 떨어지는 숫자입니다.')
    else:
        print(f'{num}은 5로 나누어 떨어지는 숫자가 아닙니다.')

```

```python
# 길이체크로 푸는 더 좋은 방법
grade, class_, number = input().split()

if len(class_) == 1 :
    class_ = '0' + class_

if len(number) == 1 :
    number = '00'+ number
elif len(number) == 2 :
    number = '0' + number

print(f'{grade}{class_}{number}')
```

```python
h, w = map(float, input().split())

# 표준몸무게
if h < 150 :
    avg_w = h - 100
elif h < 160 :
    avg_w = ((h - 150)/2) +50
else:
    avg_w = (h - 100)*0.9

# 비만도
fat = (w - avg_w) * (100/avg_w)

# 비만도에 따른 등급 판정
if fat <= 10 :
    print('정상')
elif fat <= 20:
    print('과체중')
else:
    print('비만')
```
